## Distribusi Fitur

### Fitur Kategorikal

```python
plt.figure(figsize=(18,10))

#Visualisasi Gender
plt.subplot(2,3,1)
sns.countplot(x='Gender', palette='Set2', data=data,)
plt.title('Distribusi Gender')

#visualisasi Married
plt.subplot(2,3,2)
sns.countplot(x='Married', palette='Set2',data=data)
plt.title('Distribusi Status Pernikahan')

plt.subplot(2,3,3)
sns.countplot(x='Education', palette='Set2', data=data)
plt.title('Distribusi Tingkat Pendidikan')

plt.subplot(2,3,4)
sns.countplot(x='Self_Employed', palette='Set2',data=data)
plt.title('Distribusi Self Employed')

plt.subplot(2,3,5)
sns.countplot(x='Property_Area', palette='Set2',data=data)
plt.title('Distribusi Area Properti')

plt.subplot(2,3,6)
sns.countplot(x='Loan_Status',palette='Set2' ,data=data)
plt.title('Distribusi Status Pinjaman')

plt.tight_layout()
plt.show()
```

Dari hasil distribusi data, mayoritas pemohon pinjaman adalah pria (489 pemohon), dan sebagian besar sudah menikah (398 pemohon). Kebanyakan dari mereka adalah lulusan sarjana (480 pemohon) dan bukan pekerja mandiri (500 pemohon). Pemohon berasal dari berbagai area properti dengan distribusi yang cukup merata, dengan sedikit dominasi dari area semiurban (233 pemohon). Dalam hal status pinjaman, sekitar 69% pemohon berhasil mendapatkan persetujuan pinjaman, menunjukkan bahwa mayoritas pemohon memenuhi kriteria kelayakan yang ditetapkan oleh perusahaan. 

Dari sini kita mengasumsikan beberapa hal seperti "akan terjadi sedikit bias karena distribusi gender yang tidak imbang" atau "Status pernikahan bisa menjadi faktor dalam persetujuan pinjaman, bisa jadi karena yang sudah menikah bisa jadi memiliki stabilitas finansial yang lebih tinggi."

```python
pd.crosstab(data.Gender,data.Married).plot(kind="bar", stacked=True, figsize=(5,5), color=['#f64f59','#12c2e9'])
plt.title('Gender vs Married')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
```

```python
gender_married_counts = data.groupby(['Gender', 'Married']).size().unstack()
print(gender_married_counts)
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(gender_married_counts, annot=True, cmap="YlGnBu", fmt='d')
plt.title('Jumlah Individu Berdasarkan Gender dan Status Pernikahan')
plt.xlabel('Status Pernikahan')
plt.ylabel('Gender')
plt.show()
```

Pria memiliki jumlah yang lebih besar dalam kategori menikah dibandingkan dengan wanita.

```python
pd.crosstab(data.Property_Area,data.Loan_Status).plot(kind="bar", stacked=True, figsize=(5,5))
plt.title('Property Area vs Loan Status')
plt.xlabel('Property Area')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
```

Mayoriitas pinjaman yang diterima adalah individu yang tinggal/memiliki properti di area SemiUrban daripada Urban dan Rural.

## Numerikal

```python


fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Histogram untuk ApplicantIncome
axes[0].hist(data['ApplicantIncome'], bins=20, color='skyblue')
axes[0].set_title('Distribusi ApplicantIncome')
axes[0].set_xlabel('ApplicantIncome')
axes[0].set_ylabel('Frequency')

# Histogram untuk LoanAmount
axes[1].hist(data['LoanAmount'], bins=20, color='salmon')
axes[1].set_title('Distribusi LoanAmount')
axes[1].set_xlabel('LoanAmount')
axes[1].set_ylabel('Frequency')

# Histogram untuk CoapplicantIncome
axes[2].hist(data['CoapplicantIncome'], bins=20, color='lightgreen')
axes[2].set_title('Distribusi CoapplicantIncome')
axes[2].set_xlabel('CoapplicantIncome')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Data menunjukkan bahwa kebanyakan pemohon pinjaman merupakan individu dengan pendapatan rendah (kisaran kurang dari 20.000). Pinjaman yang diajukan juka mayoritas rendah hingga sedang (berkisar 100.000 - 200.000) dan kebanyakan "coAplicant" tidak memiliki pendapataan atau banyak peminjam mengajukan secara individu.

Hal ini berimplikasi pada profil risiko dari perusahaan pinjaman, di mana pemohon dengan pendapatan yang lebih rendah mungkin memiliki risiko lebih tinggi, namun juga dapat mencakup segmen pasar yang luas.

```python
sns.set(style="darkgrid")
fig, axs1 = plt.subplots(2, 2, figsize=(10, 10))

sns.violinplot(data=data, y="ApplicantIncome", ax=axs1[0, 0], color='green')
sns.violinplot(data=data, y="CoapplicantIncome", ax=axs1[0, 1], color='skyblue')
sns.violinplot(data=data, y="LoanAmount", ax=axs1[1, 0], color='orange');
```

```python
sns.countplot(x="Loan_Amount_Term", data=data, palette="rocket")
plt.show()
```

Dilihat bahwa kebanyakan peminjam akan meminjam selama 360 atau kurang lebih sampai 1 tahun.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='ApplicantIncome', y='LoanAmount', data=data)
plt.title('Hubungan antara Pendapatan Pemohon dan Jumlah Pinjaman')
plt.xlabel('Pendapatan Pemohon')
plt.ylabel('Jumlah Pinjaman')
plt.show()
```

Seperti visualisasi diatas (yang masing-masing pendapata dan jumlah pinjaman), terlihat bahwa yang sering melakukan pinjaman adalah orang yang memiliki pendpaatan cukup rendah dengan pinjaman yang cukup rendah hingga sedang.
